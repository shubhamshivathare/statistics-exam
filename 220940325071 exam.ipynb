```python
df_merged = pd.merge(df_pre , df_app , on = 'SK_ID_CURR' , how = 'inner')
```

```python
df_merged.info()
```

```python
TARGET

AMT_INCOME_TOTAL
AMT_CREDIT_x     ==>   Final credit amount on the previous application. This differs from AMT_APPLICATION in a way that the AMT_APPLICATION is the amount for which the client initially applied for, but during our approval process he could have received different amount - AMT_CREDIT

AMT_INCOME_TOTAL ==>  Income of the client
AMT_CREDIT_y     ==>  Credit amount of the loan

NAME_CONTRACT_STATUS ==> Refused/Approval

NAME_CLIENT_TYPE ==> New/Repeater


-------------------------------------

Refused-New   (New)                 

AMT_INCOME_TOTAL
AMT_CREDIT_y
-------------------------------------

Refused-Repeater   (Prev)                 

AMT_INCOME_TOTAL
AMT_CREDIT_x

Refused-Repeater   (New)                 

AMT_INCOME_TOTAL
AMT_CREDIT_y

----------------------------------------
```

```python
df_Refused_Repeater = df_merged.iloc[np.where( (df_merged.NAME_CONTRACT_STATUS == 'Refused') & (df_merged.NAME_CLIENT_TYPE == 'Repeater') & (df_merged.Num_of_Docs != 0))]
```

```python
df_approved = df_merged.iloc[np.where(df_merged.NAME_CONTRACT_STATUS == 'Approved')]
```

```python
df_rejected = df_merged.iloc[np.where(df_merged.NAME_CONTRACT_STATUS == 'Refused')]
```

```python

plt.scatter(df_approved.AMT_CREDIT_x , df_approved.AMT_INCOME_TOTAL )



plt.scatter(df_Refused_Repeater.AMT_CREDIT_y , df_Refused_Repeater.AMT_INCOME_TOTAL )

plt.xlabel('AMT_INCOME_TOTAL')
plt.ylabel('AMT_CREDIT')
plt.title('Refused Repeater')

plt.legend(['Previously Approved' , 'Currently Rejected'])
plt.show()

# The Plot below showcase that the amount of Loan which was approved previosly but got rejected currently is not based on 
# Income Amount of a client as in both the case we could observe from graph that the credit amount remains the same so there
# are other factors that influence a loan amount and not just the Amount earned by the client.
# The Loan application which was rejected could be due to the fact that Bank thought that this loan amount would go into
# default, so Income of a client is not the only factor responsible for the loan amount rejection.
```
